```python
# Import Required Libraries
import geopandas as gpd
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm

```

```python
# Load Shapefile Data
world_data=gpd.read_file(r"C:\Users\LENOVO\Data_Engineering\World-Mapping-using-Geopandas\Data\World_Countries.shp")
world_data
```

```python
# Plot the world map using the 'code' country code column

f, ax =plt.subplots(1,1,figsize=(10,10))
world_data.plot(ax=ax, column='Code', cmap='tab10')
ax.set_title('World Map', fontsize=15)
```

```python
# Plot the world map using a diverging color scheme
f, ax =plt.subplots(1,1,figsize=(10,10))
world_data.plot(ax=ax, cmap='RdYlGn')
ax.set_title('World Map in diverging pattern')
```

```python
# Plot the world map, coloring by continent
import matplotlib.pyplot as plt

f, ax = plt.subplots(1, 1, figsize=(10, 10))
world_data.plot(ax=ax, column='Continents', cmap='tab10',legend=True, legend_kwds={'loc': 'lower left'})
plt.title('World Map Colored by Continent')
plt.show()
```

```python
# # Plot the world map based on population
f, ax =plt.subplots(1,1,figsize=(10,10))
world_data.plot(ax=ax, column='Population', cmap='RdYlGn_r', legend=True)
ax.set_title('High Density Populated Country')
```

```python
# Filter out countries with non-positive population values
world_data = world_data[world_data['Population'] > 0]

# Apply logarithmic normalization to the population data
norm=LogNorm(vmin=world_data.Population.min(),
             vmax=world_data.Population.max())

# Plot the data            
f, ax =plt.subplots(1,1,figsize=(10,10))
world_data.plot(ax=ax, column='Population', cmap='RdYlGn_r', edgecolor='w', linewidth=0.2,
            norm=norm)
ax.axis('off')
ax.set_title("Population Per Country", fontsize=14)
```
